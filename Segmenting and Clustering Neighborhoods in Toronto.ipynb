### Web Scrape Wikipedia page using pandas.read_html and get neighbourhood dataframe

```python
import requests

#Link to wikipedia page from Oct 2019 - has two neighbourhoods for M5A
wiki_url = requests.get('https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=920980179').text

neighbourhood = pd.read_html(wiki_url)

neighbourhood_df = neighbourhood[0]

neighbourhood_df
```

### Drop cells where borough column has a 'Not assigned' label

```python
neighbourhood_df = neighbourhood_df[neighbourhood_df.Borough != 'Not assigned'] 

neighbourhood_df.head()
```

### Double check to make sure neighbourhood_df is still a dataframe

```python
type(neighbourhood_df)
```

### Group neighbourhoods with the same Postcodes/ Postal codes into a single row separated by a comma

```python
neighbourhood_df = neighbourhood_df.groupby(['Postcode', 'Borough'], as_index=False, sort=False).agg(', '.join)

neighbourhood_df.head()
```

### Have the neighbourhood be the same as the borough if the neighbourhood lists a 'Not assigned' value

```python
#ex. want for Postcode M7A Borough and Neighborhood to both be Queen's Park

neighbourhood_df['Neighbourhood'] = neighbourhood_df['Neighbourhood'].replace('Not assigned', neighbourhood_df.Borough)

#check if code worked

neighbourhood_df.head()
```

```python
radius = 500
LIMIT=100

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
        
        # API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return info from each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])
        
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood', 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=toronto_data['Neighbourhood'],
                                   latitudes=toronto_data['Latitude'],
                                   longitudes=toronto_data['Longitude']
                                  )
```

### One hot encoding

```python
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

toronto_onehot['Neighbourhood'] = toronto_venues['Neighbourhood'] 

#Get neighbourhood column as first column in dataframe
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

### Get size of toronto_onehot dataframe

```python
toronto_onehot.shape
```

### Get frequency of each category in each neighbourhood

```python
toronto_group = toronto_onehot.groupby('Neighbourhood').mean().reset_index()
toronto_group
```

### Sort venues in descending order and get 5 top venues per neighbourhood

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]


#get 5 top venues per neighbourhood
num_top_venues = 5

indicators = ['st', 'nd', 'rd']

# create columns
columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# new dataframe
neighbourhoods_ven_sorted = pd.DataFrame(columns=columns)
neighbourhoods_ven_sorted['Neighbourhood'] = toronto_group['Neighbourhood']

for ind in np.arange(toronto_group.shape[0]):
    neighbourhoods_ven_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_group.iloc[ind, :], num_top_venues)

neighbourhoods_ven_sorted.head()
```

### Clustering neighbourhoods using K-means and drop neighbourhood column

```python
# Create 5 clusters

kclusters = 5

toronto_group_clustering = toronto_group.drop('Neighbourhood', 1)

toronto_group_clustering.head()
```

### Run K-means clustering

```python
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_group_clustering)

kmeans.labels_[0:5]
```

```python
# add clustering labels
neighbourhoods_ven_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = toronto_data

# merge toronto_group with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighbourhoods_ven_sorted.set_index('Neighbourhood'), on='Neighbourhood')

toronto_merged.head() 
```

### Visualize clusters

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=12)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighbourhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
